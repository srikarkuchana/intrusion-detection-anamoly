```python
X = df.iloc[:,0:20]
y = df.loc[:,'Protocol']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.5)
```

```python
X_train.head(50)
```

```python

from sklearn.svm import SVC
from sklearn.metrics import mean_squared_error,r2_score,accuracy_score
from sklearn.metrics import classification_report,confusion_matrix
```

```python
support=SVC()
```

```python
support.fit(X_train,y_train)
```

```python
y_pred=support.predict(X_test)
```

```python
mse=mean_squared_error(y_pred,y_test)
r2=accuracy_score(y_pred,y_test)
print(mse)
print(r2)
```

```python
print(classification_report(y_pred,y_test))
```

```python
print(confusion_matrix(y_pred,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
random = RandomForestClassifier(random_state=150)
```

```python
random.fit(X_train, y_train)
pred=random.predict(X_test)
```

```python
score=accuracy_score(y_test,pred)
```

```python
print(score)
```

```python
mse=mean_squared_error(pred,y_test)
print(mse)
```

```python
print(classification_report(pred,y_test))
```

```python
print(confusion_matrix(pred,y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree=DecisionTreeClassifier()
```

```python
tree.fit(X_train,y_train)
```

```python
y_p=tree.predict(X_test)
```

```python
score1=accuracy_score(y_test,y_p)
```

```python
score1
```

```python
mse=mean_squared_error(y_p,y_test)
print(mse)
```

```python
print(classification_report(y_p,y_test))
```

```python
print(confusion_matrix(y_p,y_test))
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
    
kmeans = KMeans(n_clusters=3)
labels = kmeans.fit_predict(X_train)
silhouette_avg = silhouette_score(X_train, labels)
print("Silhouette Score:", silhouette_avg)
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()

objects = ('Random Forest','DecisionTree','Support Vector',"KMeans")
y_pos = np.arange(len(objects))
performance = [score,score1,r2,silhouette_avg]

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Accuracy')
plt.title('RF vs DT vs SVM vs Kmeans')

plt.show()
```
